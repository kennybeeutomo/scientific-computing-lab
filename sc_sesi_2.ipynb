# Gauss Seidel

```python
import numpy as np
```

```python
A = [
    [8, 3, -3],
    [-2, -8, 5],
    [3, 5, 10]
]

B = [14, 5, -8]

diag = np.diag(np.abs(A))

off_diag = np.sum(np.abs(A), axis=1) - diag

if np.all(diag > off_diag):
    print("matrix is diagonally dominant")
else:
    print("not diagonally dominant")
```

```python
x1 = 0
x2 = 0
x3 = 0
epsilon = 0.01
converged = False

x_old = np.array([x1, x2, x3])
```

```python
print("iteration result")
print("k, x1,     x2,     x3")
for k in range(1, 50):
    x1 = (B[0] - A[0][1]*x2 - A[0][2]*x3) / A[0][0]
    x2 = (B[1] - A[1][0]*x1 - A[1][2]*x3) / A[1][1]
    x3 = (B[2] - A[2][0]*x1 - A[2][1]*x2) / A[2][2]
    x = np.array([x1, x2, x3])

    dx =  np.sqrt(np.dot(x-x_old, x-x_old))

    print("%d, %.4f, %.4f, %.4f"%(k, x1, x2, x3))

    if dx < epsilon:
        converged = True
        print("converged")
        break

    x_old = x

if not converged:
    print("not converged, please increase iterations")
```

# Modul

```python
import numpy as np
import matplotlib.pyplot as plt

Soal = [
    (
        [
            [2, 3, -1],
            [1, -1, 2],
            [3, 2, 1],
        ],

        [10, 4, 14],

        [5, 5, 5]
    ),
    (
        [
            [3, -0.1, -0.2],
            [0.1, 7, -0.3],
            [0.3, -0.2, 10]
        ],

        [7.85, -19.3, 71.4],

        [1, 2, 3]
    )
]


def gauss_seidel(A, B, init, epsilon=0.001, t=20):
    a = np.array(A)
    b = np.array(B)

    diags = np.diag(np.abs(a))
    np.fill_diagonal(a, 0)

    off_diags = np.sum(np.abs(a), axis=1)
    if not np.all(diags > off_diags):
        return False

    x_old = np.array(init, float)

    for i in range(t):
        x_new = np.array(x_old)
        for r, v in enumerate(a):
            x_new[r] = (b[r] - np.dot(x_new, v))/diags[r]

        print("Iter", i + 1, ":", x_new)

        dx = np.sqrt(np.dot(x_new - x_old, x_new - x_old))
        if dx < epsilon:
            return True

        x_old = x_new

    return False


for no, (A, B, Init) in enumerate(Soal):
    print("--- Soal", no + 1, "---")
    print("A   :", A)
    print("B   :", B)
    print("Init:", Init)

    if gauss_seidel(A, B, Init):
        print("Converged")
    else:
        print("Not Converged")

    print("Hasil dengan numpy:", np.linalg.solve(A, B))

    print()


X = np.array([5, 4, 6, 7, 5])
Y = np.array([10, 8, 12, 14, 10])


def leastSquares(X, Y):
    x = np.array(X)
    y = np.array(Y)

    A = np.vstack([x, np.ones(len(x))]).T
    Y = y[:, np.newaxis]

    return np.dot(np.dot(np.linalg.inv(np.dot(A.T, A)), A.T), Y)


b = leastSquares(X, Y)

plt.plot(X, Y, "ob")
plt.plot(X, b[0] * X + b[1], "r")

plt.xlabel("x")
plt.ylabel("y")

plt.show()
```
